```python
import os
import csv

def sample_files(directory):
    files = sorted(os.listdir(directory))
    return [files[i] for i in range(0, len(files), 3)]

image_dir = r'WildScenes\WildScenes2d\K-01\image'
label_dir = r'WildScenes\WildScenes2d\K-01\label'

sampled_images = sample_files(image_dir)
sampled_labels = sample_files(label_dir)

with open('./k01.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Image', 'Label']) 
    for image, label in zip(sampled_images, sampled_labels):
        writer.writerow([os.path.join(image_dir, image), os.path.join(label_dir, label)])
        
print(len(sampled_images))
```

```python
import csv
import os
import shutil

def create_directory(directory):
    if not os.path.exists(directory):
        os.makedirs(directory)

def copy_file(src, dst):
    shutil.copy2(src, dst)

def extract_images_and_labels(csv_file, image_dst, label_dst):
    create_directory(image_dst)
    create_directory(label_dst)

    with open(csv_file, 'r') as file:
        csv_reader = csv.reader(file)
        next(csv_reader)  

        for row in csv_reader:
            if len(row) != 2:
                print(f"Warning：row {row} have error, skip")
                continue

            image_src, label_src = row

            image_filename = os.path.basename(image_src)
            label_filename = os.path.basename(label_src)
            
            image_dst_path = os.path.join(image_dst, image_filename)
            label_dst_path = os.path.join(label_dst, label_filename)

            try:
                copy_file(image_src, image_dst_path)
                copy_file(label_src, label_dst_path)
                print(f"Successfully copied: {image_filename} and {label_filename}")
            except FileNotFoundError:
                print(f"Error: cannnot find {image_src} or {label_src}")
            except PermissionError:
                print(f"Error：No access permission {image_src} or {label_src}")
            except Exception as e:
                print(f"An error occurred during copying the file: {str(e)}")

csv_file = 'k01.csv'
image_dst = 'k01/images'
label_dst = 'k01/labels'

extract_images_and_labels(csv_file, image_dst, label_dst)

print("Processing completed")
```
